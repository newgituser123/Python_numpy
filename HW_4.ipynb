Постройте линейную диаграмму, используя массив t для координат по горизонтали,а массив f - для

#### Задание 3
С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3
включительно.<br>
Создайте массивы y1, y2, y3, y4 по следующим формулам:<br>
y1 = x** 2<br>
y2 = 2 * x + 0.5<br>
y3 = -3 * x - 1.5<br>
y4 = sin(x)<br>
Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с
названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных
графика в сетке, состоящей из двух строк и двух столбцов. <br>
В каждом графике массив x используется для координат по горизонтали.
- в левом верхнем - y1,
- в правом верхнем - y2,
- в левом нижнем - y3,
- в правом нижнем - y4.<br>
Дайте название графикам: 'График y1', 'График y2' и т.д.<br>
Для графика в левом верхнем углу установите границы по оси x от -5 до 5.<br>
Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.<br>
Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.

```python
x = np.linspace( 0, 10, 51)
```

```python
y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, sharex=True)   # ax.flatten()
fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)
                                            
ax1.plot(x, y1)
ax1.set_title("График y1")

ax2.plot(x, y2)
ax2.set_title("График y2")

ax3.plot(x, y3)
ax3.set_title("График y3")

ax4.plot(x, y4)
ax4.set_title("График y4");
```

```python
cnt = data.Class.value_counts()
```

```python
cnt.plot(kind='bar')
plt.title("Fraud vs Good: pure y")
plt.show()
```

```python
cnt.plot(kind='bar', logy=True)
plt.title("Fraud vs Good: log(y)")
plt.show()
```

```python
fig, ax1 = plt.subplots(nrows=1, ncols=1)
fig.set_size_inches(6, 6)
fig.subplots_adjust(wspace=0.3)
                                            
ax1.hist(data[data.Class==1].V1, label="Fraud", bins=20, color = 'red', alpha=0.5, density=True)
ax1.hist(data[data.Class==0].V1, label="Good", bins=20, color = 'grey', alpha=0.5, density=True)
plt.xlabel("V1")
plt.legend()
plt.show()
```
